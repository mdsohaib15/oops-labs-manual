```python
# Topic # 07
# 1. Explain and demonstrate the different types of inheritance in Python.
# 1------------single inheritance------------
class Animal:
          def __init__(self,name,breed):
                    self.name = name 
                    self.breed = breed
                    print(self.name,self.breed)
          def make_sound(self):
                    print("sound made by animal")

class Dog(Animal):
       def make_sound(self):
                    print("bark")

d = Dog("Dog","dogger")
d.make_sound()

a = Animal("cat","birman")
a.make_sound()
```

```python
# 2--------Multiple Inheritance-------
class Batsman:
          def __init__(self,name,role):
                    self.name  = name 
                    self.role = role
          def info1(self):
                    print(f"{self.name} is a  very good {self.role} batsman")

class Bowler:
          def __init__(self,name,role):
                    self.name  =name 
                    self.role = role
          def info2(self):
                    print(f"{self.name} is a very good {self.role} bowler")

class AllRounder(Batsman,Bowler):
          def __init__(self,name,role):
                super().__init__(name,role)    

p1 = Batsman("virat","Middle order")
p1.info1()

p2 = Bowler("Steyn","Fast")
p2.info2()

p3 = AllRounder("watson" , "opening")
p3.info1()
p3 = AllRounder("watson" , "death")
p3.info2()
```

```python
# 3---------Multilevel Inheritance---------------
class Animal:
          def __init__(self,name,species):
                    self.name = name 
                    self.species = species
          def showDetails(self):    # step3
                    print(f"Name : {self.name}")
                    print(f"Species: {self.species}")
class Dog(Animal):
          def __init__(self,name,breed):
                    Animal.__init__(self,name,species ="Dog")
                    self.breed = breed
          def showDetails(self): # step2
                    Animal.showDetails(self) 
                    print(f"Breed: {self.breed}")
class GoldenRetiever(Dog):
          def __init__(self,name,color):
                    self.color = color
                    Dog.__init__(self,name,breed = "GoldenRetiever")
          def showDetails(self): # step1
                    Dog.showDetails(self)
                    print(f"Color: {self.color}")

a1 = GoldenRetiever("Tommy","Black")
a1.showDetails()
```

```python
# 4--------Hirarchical Inheritance-------
class Game:
          def __init__(self,name,genre):
                    self.name = name
                    self.genre = genre
          def gameName(self):
                    print(f"Name: {self.name}")
                    print(f"Game Type: {self.genre}")

class Action(Game):
          def __init__(self,name,type):
                    Game.__init__(self,name,genre = "Action")
                    self.type = type
          def info(self):
                    Game.gameName(self)
                    print(f"Game Type: {self.type}")
class Horror(Game):
          pass
class Survival(Game):
          pass

g1 = Action("Resident evil","Story")
g1.info()
```

```python
# 5---------Hybrid Inheritance---------------
class Game:
          def __init__(self,name):
                    self.name = name
          def game1(self):
                    print(f"Name: {self.name}")
                    print("It is an Action Game")
          def game2(self):
                    print(f"Name: {self.name}")
                    print("It is an Adventural Game")


class Action(Game):
          def __init__(self,name,type):
                    Game.__init__(self,name)
                    self.type = type
          def info(self):
                    print(f"Game Type: {self.type}")
                    Game.game1(self)

class Adventure(Game):
          def __init__(self,name,type):
                    Game.__init__(self,name)
                    self.type = type
          def info(self):
                    print(f"Game Type: {self.type}")
                    Game.game2(self)

class Horror(Action,Adventure):
          def __init__(self,name,type):
                    self.type = type
                    Action.__init__(self,name , type)
                    Adventure.__init__(self,name , type)

          def info(self):
                    Action.info(self)
                    Adventure.info(self)


# g1 = Action("Pubg","Battle Royal")
# g1.info()
# g2 = Adventure("Sabnautica ","Survival")
# g2.info()
g3 = Horror("RE","Story")
g3.info()

```
